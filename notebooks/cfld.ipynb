```julia
# It is not clear how the distance matrix is computed. In the source, they say that "The
# matrix d_ij is the Euclidean distance matrix", but that does not give me the info on the
# location of the facilities, as the .txt only has the location of the customers.

# I assume that the facilities' locations are the same as the customers
euclidean_distance(x,y) = sqrt((x[1]-y[1])^2 + (x[2]-y[2])^2)
d = [euclidean_distance(locs[i], locs[j]) for i in I, j in J]

# this parameter configuration yielded the lowest runtimes in Crönert and Minner, 2024, Table 1
d_max = 40
β = 2
B = 40
CM_utility(i,j,r) = (d[i,j] <= d_max) ? a[i,r]/((d[i,j]+1)^β) : 0
u = [CM_utility(i,j,r) for i in I, j in J, r in R]
;
```

# Players

```julia
IPG.initialize_strategies = IPG.initialize_strategies_player_alone

function cfld_payoff(xp, x_others)
    xp = xp[1]
    x_other = x_others[1][1]  # two-player game
    return sum([
        # w[j] * sum(u[:,j,:] .* xp)
        w[j] * sum(u[:,j,:] .* xp) / (
            # sum(u[:,j,:] .* x_other)
            sum(u[:,j,:] .* xp) + sum(u[:,j,:] .* x_other)
        )
        for j in J
    ])
end

X1 = Model()
@variable(X1, x1[I,R], Bin)
@constraint(X1, sum(f .* x1) <= B)
@constraint(X1, [i in I], sum(x1[i,:]) <= 1)

X2 = Model()
@variable(X2, x2[I,R], Bin)
@constraint(X2, sum(f .* x2) <= B)
@constraint(X2, [i in I], sum(x2[i,:]) <= 1)

players = [
    Player(X1, [x1], BlackBoxPayoff(cfld_payoff), 1),
    Player(X2, [x2], BlackBoxPayoff(cfld_payoff), 2)
]

Σ, payoff_improvements = IPG.SGM(players, SCIP.Optimizer, max_iter=5, verbose=true)
```
